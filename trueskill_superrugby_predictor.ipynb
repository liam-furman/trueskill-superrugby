```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import datetime
```

```python
s = """
<td>
<div align="center"><a href="/force/" title="Western Force Super 15 Rugby">Western<br>
Force</br></a></div>
</td>
"""
soup = BeautifulSoup(s, 'html.parser')
print(soup.find('div').text.replace('\n',' '))
```

```python
teams_1 = []
teams_2 = []
scores_1 = []
scores_2 = []
dates = []
for year in [str(y) for y in range(2006, 2017)]:
    url = 'http://www.superxv.com/results/{year}-super-rugby-results/'.format(year=year)
    page = requests.get(url)
    if page.status_code != 200:
        print('Error downloading {url}'.format(url=url))
    else:
        print('Successfully downloaded {url}'.format(url=url))
        for table in BeautifulSoup(page.content, 'html.parser').find_all('table'):
            for row in table.find_all('tr')[1:]:
                entries = row.find_all('td')
                print('\n\nNEW ROW')
                print(row)
                if entries[0].find('div').text.strip() != 'Bye':
                    teams_1.append(entries[0].find('div').text.strip().replace('\n',' '))
                    teams_2.append(entries[3].find('div').text.strip().replace('\n',' '))
                    scores_1.append(entries[1].find('div').text.strip())
                    scores_2.append(entries[4].find('div').text.strip())
                    dates.append(entries[5].find('div').text)
```

```python
df = pd.DataFrame({
        'date':dates,
        'team_1':teams_1,
        'team_2':teams_2,
        'score_1':scores_1,
        'score_2':scores_2})
df.date = pd.to_datetime(df.date)
df
```
